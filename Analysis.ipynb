# Data cleaning
Apply some fine-grained filtering

```python
def clean(data):
    odo_max = 60000
    age_max = 5
    price_max = 16000
    exclude_brands = ['Peugeot', 'Holden', 'SKODA', 'Ford', 'Great', 'Jeep', 'Chery']
    drop_cols = ['manufacturer_marketing_year', 'title', '_type', 'id']

    initial_len = data.shape[0]
    print("Initial num of records: ", initial_len)

    data = pd.read_csv(filename)
    df = data
    df = data[data['odometer'] <= odo_max]
    df = df[df['price'] <= price_max]
    df = df[df['year'] >= 2018-age_max]
    brand_mask = df['make'].isin(exclude_brands)
    df = df[~brand_mask]
    df = df.drop(drop_cols, axis=1)

    final_len = df.shape[0]
    print("Final num of records: ", final_len)
    
    return df
```

```python
# find average price per make/model
df = clean(data)
df.groupby(['make', 'model']).agg(['mean', 'count']).sort_values(by=('price', 'mean'), ascending=True)
```

```python
# add field for price per kilometer on odometer (non-agg)
df = clean(data)
df['pricepkm'] = df['price'] / df['odometer']
df.sort_values(['pricepkm'])
```

```python
# work out value sum (non-agg)
from sklearn.preprocessing import MinMaxScaler

df = clean(data)
df['age'] = 2018 - df['year']

scaler = MinMaxScaler() 
normalize = ['age', 'price', 'odometer', 'engine_capacity']
normalized = [name + '_norm' for name in normalize]
df[normalized] = df[normalize]
df[normalized] = scaler.fit_transform(df[normalized])
df

df['value'] = 0.3*df['price_norm'] + 0.3*df['odometer_norm']+ 0.2*df['age_norm'] + 0.2/df['engine_capacity_norm']
df.sort_values(['value'], ascending=True)
df.drop(normalized, axis=1)
pd.set_option('display.max_colwidth', -1)
df
```
