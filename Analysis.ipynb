# Data cleaning
Apply some fine-grained filtering

```python
odo_max = 60000
age_max = 5
price_max = 16000
exclude_brands = ['Peugeot', 'Holden', 'SKODA', 'Ford', 'Great', 'Jeep']
drop_cols = ['manufacturer_marketing_year', 'title', '_type', 'id']

initial_len = data.shape[0]
print("Initial num of records: ", initial_len)

data = pd.read_csv('data/items_carsales_13.csv')
df = data
df = data[data['odometer'] <= odo_max]
df = df[df['price'] <= price_max]
df = df[df['year'] >= 2018-age_max]
brand_mask = df['make'].isin(exclude_brands)
df = df[~brand_mask]
df = df.drop(drop_cols, axis=1)

final_len = df.shape[0]
print("Final num of records: ", final_len)
```

```python
# find average price per make/model
df.groupby(['make', 'model']).agg(['mean', 'count']).sort_values(by=('price', 'mean'), ascending=True)
```

```python
# add field for price per kilometer on odometer (non-agg)
df['pricepkm'] = df['price'] / df['odometer']
df.sort_values(['pricepkm'])
```

```python
# work out value sum (non-agg)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(df) 
df.loc[:,:] = scaled_values

df['age'] = 2018 - df['year']
df['value'] = df['price'] + df['odometer']+ df['age'] + df['engine_capacity']
df
```
